```python
import pandas as pd
from pathlib import Path
data_root = Path('.')
train_source = pd.read_parquet(data_root / 'label_train_source.parquet')
train_target = pd.read_parquet(data_root / 'label_train_target.parquet')
test_source = pd.read_parquet(data_root / 'label_test_source.parquet')
meta_song = pd.read_parquet(data_root / 'meta_song.parquet')
```

```python
train_X = train_source.pivot_table(index='session_id', columns='listening_order', values='song_id', aggfunc='first', fill_value=0)
train_X
```

```python
train_Y = train_target.pivot_table(index='session_id', columns='listening_order', values='song_id', aggfunc='first', fill_value=0)
train_Y
```

```python
from sklearn.preprocessing import LabelEncoder

# 創建一個LabelEncoder
label_encoder = LabelEncoder()

# 將每一列進行label encoding
for column in train_X.columns:
    if train_X[column].dtype == 'object':
        train_X[column] = label_encoder.fit_transform(train_X[column])

# 重複上述步驟以處理Test_X，確保使用相同的LabelEncoder對象
for column in train_Y.columns:
    if train_Y[column].dtype == 'object':
        train_Y[column] = label_encoder.fit_transform(train_Y[column])
# 接下來，您可以使用Train_X和Train_Y來訓練您的模型


```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
# 假設 Train_X 和 Train_Y 是 pandas 的 DataFrame
X_train, X_test, y_train, y_test = train_test_split(train_X.values, train_Y.values, test_size=0.1, random_state=42)
```

```python
# 建立線性回歸模型
model = RandomForestClassifier(verbose=1)
# 訓練模型
model.fit(X_train, y_train)
```

```python
# 在測試集上進行預測
predictions = model.predict(X_test)
```

```python
# 之後，如果您需要將預測轉換回原始字串值，可以使用inverse_transform方法
inverse_transformed_values = label_encoder.inverse_transform(predictions)
```

```python
prediction = pd.DataFrame(inverse_transformed_values)
prediction.columns = [21,22,23,24,25]  #自訂欄位名稱
prediction
```

```python
y_test = pd.DataFrame(y_test)
y_test.columns = [21,22,23,24,25]  #自訂欄位名稱
y_test
```

```python
# 將 inverse_transformed_values 與 y_test 進行比較
r = prediction[['21', '22', '23', '24', '25']] == y_test[['21', '22', '23', '24', '25']].values
r['25'] = r['25'] & ~(r[['21', '22', '23', '24']].any(axis=1))
r['24'] = r['24'] & ~(r[['21', '22', '23']].any(axis=1))
r['23'] = r['23'] & ~(r[['21', '22']].any(axis=1))
r['22'] = r['22'] & ~(r[['21']].any(axis=1))
```

```python
import rich
w = [1., 0.63, 0.5, 0.43, 0.38]
all_songs = meta_song.shape[0]
dcg = r.mean() @ w
predict_songs = y_test[['21', '22', '23', '24', '25']].sample(frac=0.25).nunique()
fill_songs = 4*test_source['session_id'].nunique()
coverage = (predict_songs+fill_songs)/all_songs
rich.print({'dcg':dcg, 'coverage':coverage, 'total':dcg*0.8+coverage*0.2})
```
